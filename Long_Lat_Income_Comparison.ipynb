```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import random
%matplotlib notebook
from numpy.polynomial.polynomial import polyfit

file = "kaggleincome.csv"
df_main = pd.read_csv(file,engine = "python")
df_main.head()

```

```python
df_main = df_main[(df_main["Lat"] > 25) & (df_main["Lat"] < 50)]
df_main = df_main.round({'Lat': 0, 'Lon': 0})
df_main["Number of Households"] = df_main["sum_w"] / df_main["Mean"]
df_main_grpby_lat = df_main.groupby(["Lat"]).sum()
df_main_grpby_lat["Mean Income"] = df_main_grpby_lat["sum_w"] / df_main_grpby_lat["Number of Households"]
df_main_grpby_lat.plot( y=["Mean Income"],kind="line")

x = ("Lat")
y = ("Latitude")

# Fit with polyfit
#b, m = polyfit(x, y, 1)
#plt.plot(x, y, '.')
#plt.plot(x, b + m * x, '-')
#plt.show()


plt.title("Income vs Lattitude")
plt.ylabel("Mean Income")
plt.xlabel("Latitude")
plt.grid(True)
plt.show()
```
