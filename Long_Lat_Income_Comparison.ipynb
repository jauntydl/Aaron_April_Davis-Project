```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
import random
%matplotlib notebook

file = "kaggleincome.csv"
df_main = pd.read_csv(file,engine = "python")
df_main.head()

```

```python
df_main = df_main[(df_main["Lat"] > 25) & (df_main["Lat"] < 50)]
df_main = df_main.round({'Lat': 0, 'Lon': 0})
df_main["Number of Households"] = df_main["sum_w"] / df_main["Mean"]
df_main_grpby_lat = df_main.groupby(["Lat"]).sum()
df_main_grpby_lat["Mean Income"] = df_main_grpby_lat["sum_w"] / df_main_grpby_lat["Number of Households"]
df_main_grpby_lat.plot( y=["Mean Income"],kind="line")

plt.title("Income vs Longitude")
plt.ylabel("Mean Income")
plt.xlabel("Lattitude")
plt.grid(True)
plt.show()
```

```python
df_main = df_main[(df_main["Lon"] > -10.0) & (df_main["Lon"] < -90.0)]
df_main = df_main.round({'Lat': 0, 'Lon': 0})
df_main["Number of Households"] = df_main["sum_w"] / df_main["Mean"]
df_main_grpby_lon = df_main.groupby(["Lon"]).sum()
df_main_grpby_lon["Mean Income"] = df_main_grpby_lon["sum_w"] / df_main_grpby_lon["Number of Households"]
df_main_grpby_lon.plot( y=["Mean Income"],kind="line")
```
